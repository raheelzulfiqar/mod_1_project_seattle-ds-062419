```python
#scale the data logs
```

```python
data_log_kc_data_2012.head()
```

```python
Sqft = data_log_kc_data_2012["SqFtTotLiving"]
BldgGrade = data_log_kc_data_2012['BldgGrade']

logyrbuilt = data_log_kc_data_2012["logYrbuilt"]
logbedrooms = data_log_kc_data_2012["logBedrooms"]
logbathfull = data_log_kc_data_2012["logBathFullCount"]
```

```python
scaled_Sqft = (Sqft-min(Sqft))/(max(Sqft)-min(Sqft))
scaled_BldgGrade = (BldgGrade-min(BldgGrade))/(max(BldgGrade)-min(BldgGrade))

scaled_Yrbuilt = (logyrbuilt-np.mean(logyrbuilt))/np.sqrt(np.var(logyrbuilt))
scaled_bedrooms = (logbedrooms-np.mean(logbedrooms))/(max(logbedrooms)-min(logbedrooms))
scaled_Bathfull= (logbathfull-np.mean(logbathfull))/np.sqrt(np.var(logbathfull))
```

```python
data_cont_scaled = pd.DataFrame([])

data_cont_scaled["Sqft"] = scaled_Sqft
data_cont_scaled["BldgGrade"]= scaled_BldgGrade
data_cont_scaled["Yrbuilt"]= scaled_Yrbuilt
data_cont_scaled["Bedrooms"] = scaled_bedrooms
data_cont_scaled["BathfullCount"] = scaled_Bathfull

data_cont_scaled.hist(figsize  = [6, 6]);
```

```python
data_cont_scaled.dropna(inplace=True)
data_cont_scaled.head()
```
